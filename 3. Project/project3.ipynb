### Step 3

```python
fertility_2015 = df_fertility['2015']
expectancy_2015 = df_expectancy['2015']

fertility_expectancy_2015 = pd.merge(fertility_2015, expectancy_2015, on='country')
fertility_expectancy_2015.columns = ['fertility', 'life_expectancy']
fertility_expectancy_2015.head()
```

### Step 4

```python
fertility_expectancy_2015_clean = fertility_expectancy_2015.dropna()
fertility_expectancy_2015_clean.head()
```

### Step 5

```python
x=fertility_expectancy_2015_clean['fertility']
y=fertility_expectancy_2015_clean['life_expectancy']
plt.scatter((x), (y))
plt.xlabel('Fertility')
plt.ylabel('Life Expectancy')
plt.title('Fertility vs Life Expectancy in 2015')
plt.axis([0, 9, 25, 90])
plt.show()
```

### Step 7

```python
def scatterplot_generator(year):
    fertility = df_fertility[year]
    expectancy = df_expectancy[year]

    fertility_expectancy = pd.merge(fertility, expectancy, on='country')
    fertility_expectancy.columns = ['fertility', 'life_expectancy']

    fertility_expectancy_clean = fertility_expectancy.dropna()

    x=fertility_expectancy_clean['fertility']
    y=fertility_expectancy_clean['life_expectancy']
    plt.scatter((x), (y))
    plt.xlabel('Fertility')
    plt.ylabel('Life Expectancy')
    plt.title('Fertility vs Life Expectancy in ' + year)
    plt.axis([0, 10, 20, 100])
    return plt
```

### Step 8

```python
for year in range(1960, 2011):
    plot = scatterplot_generator(str(year))
    plot.savefig('images/fertility_vs_life_expectancy_in_' + str(year) + '.png')
    plot.clf()
```

### Step 10

```python
df_population.dropna(inplace=True)

def pop_convert(x):
    if isinstance(x, (int,float)):
        return int(x)
    if x.endswith('k'):
        result = float(x[:-1]) * 1000
        return int(result)
    elif x.endswith('M'):
        result = float(x[:-1]) * 1000000
        return int(result)
    elif x.endswith('B'):
        result = float(x[:-1]) * 1000000000
        return int(result)
    else:
        return int(x)

df_population= df_population.map(pop_convert)
df_population.head()
```

### Step 11

```python
def devide_as_needed(x):
    return x/100000

df_population = df_population.map(devide_as_needed)

fertility = df_fertility['2015']
expectancy = df_expectancy['2015']
population = df_population['2015']

fertility_expectancy = pd.merge(fertility, expectancy, on='country')
fertility_expectancy_population = pd.merge(fertility_expectancy, population, on='country')
fertility_expectancy_population.columns = ['fertility', 'life_expectancy', 'population']

fertility_expectancy_population_clean = fertility_expectancy_population.dropna()

x=fertility_expectancy_population_clean['fertility']
y=fertility_expectancy_population_clean['life_expectancy']
s=fertility_expectancy_population_clean['population']

plt.scatter((x), (y), s=s)
plt.xlabel('Fertility')
plt.ylabel('Life Expectancy')
plt.title('Fertility vs Life Expectancy in 2015')
plt.axis([0, 10, 20, 100])
plt.show()
```

### Step 12

```python
df_continents = pd.read_csv('continents.csv', sep=';')
df_continents.head()

fertility_expectancy_population_continent = pd.merge(fertility_expectancy_population, df_continents, on='country')
fertility_expectancy_population_continent.dropna(inplace=True)

fertility_expectancy_population_continent_clean = fertility_expectancy_population_continent.dropna()

x=fertility_expectancy_population_continent_clean['fertility']
y=fertility_expectancy_population_continent_clean['life_expectancy']
s=fertility_expectancy_population_continent_clean['population']
color_guide = fertility_expectancy_population_continent_clean['continent']

sns.scatterplot(x=x, y=y, size=s, hue=color_guide, sizes=(20, 500), legend=False)
plt.xlabel('Fertility')
plt.ylabel('Life Expectancy')
plt.title('Fertility vs Life Expectancy in 2015')
plt.axis([0, 10, 20, 100])
plt.show()
```
